```python
# Libraries needed for the lab are imported.
import pandas as pd                     # Used for creating and handling dataframes.i
import numpy as np                      # Used for math operations.
import seaborn as sns                   # Used for statistical methods and data visualization. 
from matplotlib import pyplot as plt    # Used for data visualization as well.
```

```python
# Data from the repository is imported.
dataUrl = "https://raw.githubusercontent.com/corredor-john/ExploratoryDataAnalisys/main/Varios/auto_data.csv"

# Dataframe is created with the previous URL.
df_autos01 = pd.read_csv(dataUrl)

# It's required to print the first 5 rows/records.
df_autos01.head(10)
```

```python
# It's also required to show the last 10 rows/records.
df_autos01.tail(10)
```

```python
# General information of the dataframe is shown.
df_autos01.info()
```

```python
# With the previous information obtained, we can see that column 9, "Market Category", only has filled 8172 cells out of the possible 11914 .This means that more than 3,500 records are missing this information. Because of this, it is decided to erase from the dataframe the entire column.
df_autos02 = df_autos01.drop(['Market Category'], axis= 1)

# First 10 registers are shown again.
df_autos02.head(10)
```

```python
# Columns 4,5 and 8 ("Engine HP", "Engine Cylinders", "Number of doors"), have missing or impossible data. With the objective of preserving as much data as possible, and also noting that:
#  - Missing records don't represent a big number of total records.
#  - The type of the columns mentioned is float
# data imputation will be carried out. This under the risk of imbalancing or biasing the data.    
df_autos02['Engine HP'] = df_autos02['Engine HP'].fillna(df_autos02['Engine HP'].mean())
df_autos02['Engine Cylinders'] = df_autos02['Engine Cylinders'].fillna(df_autos02['Engine Cylinders'].mean())
df_autos02['Number of Doors'].fillna(df_autos02['Number of Doors'].mean(), inplace= True)

# Information of the dataframe is shown to validate. 
df_autos02.info()
```

```python
# Now the only remaining column with missing data is "Engine fuel type". Given that we want to achieve uniformity and that the column is only missing 3 records, we erase said records.
df_autos03 = df_autos02.dropna()

df_autos03.info()
```

```python
# We know erase duplicates from the now balanced dataframe.
df_autos04 = df_autos03.drop_duplicates()

# Information of the new dataframe is shown.
df_autos04.info()
```

```python
# We know change the name of some columns for the sake of learning and also readability (to some extent). Also it's recommended to remove spaces from columns names.
df_autos05 = df_autos04.rename(columns={
                                    'Engine Fuel Type': "Fuel_Type", 
                                    'Engine HP': "Eng_HP",
                                    'Engine Cylinders': "Eng_Cylinders", 
                                    'Transmission Type': "Transm_Type", 
                                    'Number of Doors': "Num_Doors", 
                                    'Vehicle Size': "Size", 
                                    'Vehicle Style': 'Style',
                                    'highway MPG': "Highway_MPG",
                                    'city mpg': "City_MPG",
                                    'MSRP': 'Price'
                                    } 
                            )

# Visual verification of renaming process.
df_autos05.head()
```

```python
# Now that data is pretty and uniform, we make an statistical summary of numerical data.
df_autos05.describe()
```

```python
# Next, we want to see if there's outliers in the price column. We'll use seaborn for the visualization, and a box plot for the plot.

sns.boxplot(x=df_autos05['Price'])
```

## Estudio de Outliers
Los outliers son valores que son diferentes ... (i have a photo of the text)

```python
# We now study the outliers in the folowwing columns (Price, City_MPG, )
sns.boxplot(x = df_autos05['Price'])
plt.show()
```

```python
# Values of quartiles are now defined
q1 = df_autos05.quantile(0.25)
q3 = df_autos05.quantile(0.75)

# The interquartile range is defined (IRQ)
irq = q3 - q1

# We create a new dataframe by discarting outliers or atypical values.
df_autos06 = df_autos05[~((df_autos05 < (q1 - 1.5 * irq)) | (df_autos05 > (q3 + 1.5 * irq))).any(axis = 1)]

# We visualize the changes by showing the info of the dataframe created.
df_autos06.info()
```
